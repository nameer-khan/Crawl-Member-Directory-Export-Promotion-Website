```python
from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome() 
driver.get("https://epch.in/index.php?option=com_jumi&fileid=4&Itemid=162&page=1")
```

```python
df = pd.DataFrame(columns=['Name','Address','Tel','Email','Contact Persons','Products'])
```

```python
df
```

```python
columns = list(df)
data=[]




for i in range(1,1059):      
    wait = WebDriverWait(driver, 10)
    url= "https://epch.in/index.php?option=com_jumi&fileid=4&Itemid=162&page="+str(i)
    driver.get(url)
    j=3
    while(j<=75):
        names=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td""").format(j-1))
        for name in names:
            name=name.text
        address1=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td""").format(j))
        for add1 in address1:
            add1=add1.text
        address2=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td""").format(j+1))
        for add2 in address2:
            add2=add2.text
        address=add1+"\n"+add2
        tel=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td[2]""").format(j+2))
        for telephone in tel:
            telephone=telephone.text
        email=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td[2]/a""").format(j+3))
        for em in email:
            em=em.text
        contact=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td[2]""").format(j+4))
        for con in contact:
            con=con.text
        product=driver.find_elements_by_xpath(("""//*[@id="content"]/div/div[2]/table/tbody/tr[{}]/td[2]""").format(j+5))
        for pro in product:
            pro=pro.text
        j=j+8;
        values=[name,address,telephone,em,con,pro]
        zipped =zip(columns,values)
        a_dictionary = dict(zipped)
        data.append(a_dictionary)
    
    
```

```python


df = df.append(data, True)
```
